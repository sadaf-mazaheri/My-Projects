```python
df = pd.read_csv('retail.dat.gz',compression='gzip' , sep=',')
df
```

```python
df.loc[88161] = df.columns  # adding a row
df = df.rename(columns={df.columns[0] : 'items' })
df
```

```python
df['items'] = df['items'].apply(lambda x : x.split(' '))
```

```python
count_1_item = {}

def count_1 (x) :
    for i in x:
        if i not in count_1_item.keys() : 
            count_1_item[i] = 1
        else :
            count_1_item[i] += 1

df['items'].apply(count_1)

del count_1_item[""]
```

```python
data = pd.DataFrame()
data['items']  = count_1_item.keys()
data['number']  = count_1_item.values()
data
```
